```python
from langchain_openai import ChatOpenAI
from langchain.tools import tool
from ddgs import DDGS
import requests
import json
```

**üß© STEP 3 ‚Äî Web Search Tool (Free)**
‚úî DuckDuckGo (no API key)

```python
@tool
def web_search(query: str):
    """Search the web for live information."""
    results = DDGS().text(query, max_results=5)
    return results
```

**üß© STEP 4 ‚Äî Generic API Call Tool**

This lets your agent call ANY API.

```python
def call_api(url: str):
  #call nme
  try:
    resp = requests.get(url)
    return resp.json()
  except:
       return "API call failed"
```

**#üß© STEP 5 ‚Äî Add Tools to Your Tools List**

```python
tools = {
    "web_search": web_search,
    "call_api": call_api
}
```

**üß© STEP 6 ‚Äî Update the Execution Agent Prompt**

‚úÖ What is executor_prompt?

It is instructions for your AI agent.

- an agent needs rules:

- When to use tools

- How to pick a tool

- What format to output

- What to ignore

- What style to follow

This prompt is simply the brain rules for your ‚ÄúExecution Agent‚Äù.

      Given this step:
      {step}


‚û°Ô∏è When the AI receives a step like:

"Find price of BTC"

We insert it into {step}.

Choose ONE tool and return:

      {
      "tool": "...",
      "args": { ... }
      }


```python
executor_prompt = """
You are an Execution Agent.

Available tools:
- web_search(query)
- call_api(url)
- add_numbers(a,b)
- ai_fact(topic)

Given this step:
{step}

Choose ONE tool and return:
{
  "tool": "...",
  "args": { ... }
}
"""
```
