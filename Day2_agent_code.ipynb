```python
!pip install -q langchain langchain-community langchain-core python-dotenv openai

from langchain.agents import initialize_agent, AgentType, load_tools
from langchain_openai import ChatOpenAI
import os

os.environ["OPENAI_API_KEY"] = "Add Your API Key Here"  #Add api key here
os.environ["OPENAI_API_BASE"] = "https://openrouter.ai/api/v1"

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)
tools = load_tools(["llm-math"], llm=llm)
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

query = "What is 29% of 450?"    # test it
response = agent.run(query)
print("\n Final Answer:", response)
```
